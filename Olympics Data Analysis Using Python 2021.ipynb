```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Chargement des donnees

atheles = pd.read_csv("/home/aboubakar/projects/Olympics_Data_Analysis_Using_Python_2021/data/athlete_events.csv")
regions = pd.read_csv("/home/aboubakar/projects/Olympics_Data_Analysis_Using_Python_2021/data/noc_regions.csv")
```

```python
atheles.head()
```

```python
regions.head()
```

```python
#jointure des deux dataframes

atheles_df = atheles.merge(regions, how = 'left', on = 'NOC')
atheles_df.head()
```

```python
atheles_df.shape
```

```python
#Modification de nom de colonnes

atheles_df.rename(columns = {'region':'Region', 'notes':'Notes'}, inplace=True)
```

```python
# Top des pays particiapants
top_10_countries = atheles_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
#Affichage des top 10 sous forme de grap

plt.figure(figsize=(12, 6))
#plt.xtricks(rotation=20)
plt.title("Overall Participation by country")
sns.barplot(x=top_10_countries.index, y=top_10_countries, palette='Set2');
```

```python
#Répartition par âge des participants
plt.figure(figsize=(12, 6))
plt.title("Age distribution of the aatheles")
plt.xlabel('Age')
plt.ylabel('Number of participant')
plt.hist(atheles_df.Age, bins = np.arange(10,80,2), color = 'orange', edgecolor = 'white');
```

```python
#Participants homme et femme
gender_counts = atheles_df.Sex.value_counts()
gender_counts
```

```python
#Affichage graphique des participants hommes et femmes
plt.figure(figsize=(12, 6))
plt.title('Gender Distribution')
plt.pie(gender_counts, labels=gender_counts.index, autopct = '%1.1f%%', startangle = 150, shadow = True);
```

```python
#Médaille totale
atheles_df.Medal.value_counts()
```

```python
#Nombre total d'athlètes féminines dans chaque olympiade
female_participants = atheles_df[(atheles_df.Sex == 'F') & (atheles_df.Season == 'Summer')][['Sex', 'Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.tail()
```

```python
womenOlumpics = atheles_df[(atheles_df.Sex == 'F') & (atheles_df.Season == 'Summer')]
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
sns.countplot(x='Year', data=womenOlumpics, palette='Spectral')
plt.title('Women Participant')
```

```python
part = womenOlumpics.groupby('Year')['Sex'].value_counts()
plt.figure(figsize=(20,10))
part.loc[:, 'F'].plot()
plt.title('Plot of female atheles over time')
```

```python
#Atheles ayant la medialle en or
goldMedals = atheles_df[(atheles_df.Medal == 'Gold')]
goldMedals.head()
```

```python
#prendre que les valeurs qui sont différentes de NaN
goldMedals = goldMedals[np.isfinite(goldMedals['Age'])]
```

```python
#or au-delà de 60 ans
goldMedals['ID'][goldMedals['Age']>60].count()
```

```python
sportig_event = goldMedals['Sport'][goldMedals['Age']>60]
sportig_event
```

```python
#Affichage de resultat sous form graphique
plt.figure(figsize=(10,5))
plt.tight_layout()
sns.countplot(sportig_event)
plt.title("Golds Medals for Atheles over 60 years")
```

```python
#Médailles en or pour chaque pays
goldMedals.Region.value_counts().reset_index(name='Medal').head(5)
```

```python
totalGoldMedals = goldMedals.Region.value_counts().reset_index(name='Medal').head(6)
g = sns.catplot(x="index", y="Medal", data=totalGoldMedals,
               height=5, kind='bar', palette='rocket')
g.despine(left=True)
g.set_xlabels('Top 5 countries')
g.set_ylabels('Number of Medals')
plt.title('Gold Medals per country')
```

```python
#Olympic de Rio
max_year = atheles_df.Year.max()
print(max_year)

team_names = atheles_df[(atheles_df.Year == max_year) & (atheles_df.Medal == 'Gold')].Team

team_names.value_counts().head(10)
```

```python
sns.barplot(x=team_names.value_counts().head(20), y=team_names.value_counts().head(20).index)

plt.ylabel(None);
plt.xlabel('Countrywise Medals for the year 2016')
```
